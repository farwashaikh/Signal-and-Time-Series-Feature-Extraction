```python
from scipy.signal import find_peaks
import numpy as np
import matplotlib.pyplot as plt

# Simulated ECG data
ecg = np.sin(np.linspace(0, 10, 100)) + np.random.normal(0, 0.1, 100)

# Find peaks
peaks, _ = find_peaks(ecg, height=0.5)

# Plot
plt.plot(ecg)
plt.plot(peaks, ecg[peaks], "x")
plt.title('ECG Signal with Peaks')
plt.show()
```
